```python
import pandas 
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pandas.read_csv("../data/CC26Slack.csv" , parse_dates=["Date"])
```

```python
# Date and daily active members
daily_members_average = data[["Date","Total enabled membership","Daily active members","Daily members posting messages","Messages posted"]]
plt.figure(figsize=(25,3))
plt.plot(daily_members_average["Date"],daily_members_average["Daily active members"])
```

```python
# Average users per week
data["Weekday"] = data["Date"].dt.strftime("%A")
data["Daynumber"] = data["Date"].dt.strftime("%w")
weekday_avrg = data.groupby(["Weekday","Daynumber"]).mean()
weekday_avrg = weekday_avrg.sort_values("Daynumber")

plt.figure(figsize=[15,5])
plt.plot(weekday_avrg.index.get_level_values(0), weekday_avrg["Daily active members"])
```

```python
plt.figure(figsize=[15,5])
#plot(xAxis,yAxis)
plt.plot(data["Date"],data["Messages in public channels"], label = 'Messages in public channels', linestyle="--")
plt.plot(data["Date"],data["Messages in DMs"], label = 'Messages in DMs', linestyle="-")
plt.legend()
```
